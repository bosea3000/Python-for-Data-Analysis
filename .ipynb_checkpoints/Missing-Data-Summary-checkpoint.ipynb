```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_excel('SampleData-2.xls')
df.head()
```

```python
def missingSummary(data):
    total = data.isnull().sum()
    percent = (data.isnull().sum() / data.isnull().count()) * 100
    missingDf = round(pd.concat([total, percent], axis=1, keys=['Total', 'Percent']),0).sort_values('Percent', 
                                                                                                    ascending=False)
    missingDf = missingDf.loc[missingDf['Percent'] > 0]
    missingDf.index.name = 'Attributes'
    missingDf.columns.name = 'Missing-Values'
    print('df-Complete\n')
    
    plt.figure(figsize=(8,8))
    sns.set_style("darkgrid")
    ax = sns.barplot(x=missingDf.Percent, y=missingDf.index, data=missingDf)
    
    plt.xlabel('Percent Missing')
    plt.ylabel('Attributes')
    plt.title('Missing Data Summary')
    
    plt.tight_layout()
    print('Print-Complete')
    plt.show()
    
    return missingDf.sort_values('Percent', ascending=False)
```

```python
missingdf = missingSummary(df)
missingdf
```

```python
def cleanedFrame(rawdata, missingdf, col_cutoff, row_cutoff):
    cols_to_drop = list(missingdf.loc[missingdf['Percent'] >= col_cutoff].index)
    rows_to_drop = list(missingdf.loc[missingdf['Percent'] <= row_cutoff].index)
    
    cleanedDf = rawdata.copy()
    cleanedDf = cleanedDf.drop(cols_to_drop, axis=1)
    cleanedDf = cleanedDf.dropna(subset=rows_to_drop, how='any')
    cleanedDf.reindex(np.arange(len(cleanedDf)))
    
#     if knn == 'Y':
#         knn_num = input('How many nearest neighbors?')
#     else: 

    cleanedDf_missingDf = missingSummary(cleanedDf)
    return cleanedDf, cleanedDf_missingDf
```

```python
cleanedDf, cleaned_missingDf = cleanedFrame(df, missingdf, 80, 5)
```

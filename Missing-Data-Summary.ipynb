```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_excel('SampleData-2.xls')
df.head()
```

```python
def missingSummary(data):
    total = data.isnull().sum()
    percent = (data.isnull().sum() / data.isnull().count()) * 100
    missingDf = round(pd.concat([total, percent], axis=1, keys=['Total', 'Percent']),0).sort_values('Percent', 
                                                                                                    ascending=False)
    print('df-Complete\n')
    
    plt.figure(figsize=(8,8))
    sns.set_style("darkgrid")
    ax = sns.barplot(x=missingDf.Percent, y=missingDf.index, data=missingDf)
    
    plt.xlabel('Percent Missing')
    plt.ylabel('Attributes')
    plt.title('Missing Data Summary')

    print('Print-Complete')
    plt.tight_layout()
    plt.show()
   
    return missingDf.sort_values('Percent', ascending=False)
```

```python
missingSummary(df)
```

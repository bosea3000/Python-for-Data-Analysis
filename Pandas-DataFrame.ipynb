```python
#Word-Count with DataFrame
def wordCounter(text):
    cwd = os.getcwd()
    file = open(cwd + '/' + text, 'r')
    
    counter={}
    for word in file.read().split():
        if word not in counter:
            counter[word] = 1
        else:
            counter[word] += 1
    
    df4 = Series(counter).to_frame()
    df4.columns = ['Count']
    
    return df4.sort_values('Count', ascending=False)
```
